### 10. 반복문
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문
- While()
- for() 

<br>반복하는 횟수가 정해져 있는 경우, 명확한 경우에 for문을 사용. 
<br>그리고 얼마나 반복 수행해야 할 지 모를 때, 특정한 조건에 도달할때까지 계속해 명령문을 반복해야 할 때 while문을 사용

```python
# while 문 사용
num=0;
while num < 13:
    num += 1
    print("%d 파이썬 최고" %num)
```

```python
# while 문 무한루프 반복문
while True:
    print("무한루프")
    break # 반복문을 바로 빠져나올 때 사용.
```

```python
number = 1
while True: # 무조건 실행
    print(number) 
    number += 1 # number에 +1 해서 number에 저장
    if number > 3: # number가 3보다 크면 
        break # 중단
```

```python
# 두 정수를 입력받아 반복문을 활용해 두 정수의 합 출력하기
while True:
    num1 = int(input("첫 번째 정수 입력 : "))
    num2 = int(input("두 번째 정수 입력 : "))

    if num1==0 and num2==0 :
        print("프로그램이 종료되었습니다.")
        break
    else :
        num1 + num2
        print("두 정수의 합 : %d" %(num1+num2))
    break
```

```python
# 현재 몸무게와 목표 몸무게를 입력받고 주차별 감량 몸무게를 입력하면 목표몸무게 달성시 종료하는 알고리즘 
weight = int(input("현재 몸무게 : "))
goal = int(input("목표 몸무게 : "))
week = 1

while True:
    week_weight = int(input("%d주차 감량 몸무게 : " %week))
    week += 1 # week를 +1씩 증가시키고 다시 저장
    weight -= week_weight # 현재 몸무게에서 주차별 몸무게를 빼고 뺀 값을 현재 몸무게에 다시 저장
    if weight <= goal: # 현재 몸무게가 목표 몸무게보다 작거나 같을 때
        break # 중단
print("%d kg 달성!! 축하합니다" %weight)
```

### 14-1. replace(이걸, 이렇게)
- 괄호의 왼쪽에 들어갈 문자열을, 오른쪽 문자열로 바꿔주는 함수  

```python
review_list1 = ["윌 스미스가 하드캐리ㅋㅋㅋㅋㅋㅋㅋ", 
                "자스민 너무 멋지고 팬 엄청나게 생길듯"]
review_list2 = ["색감도 노래도 너무 화려하고 재밌었어요ㅋㅋㅋ", 
                "오늘부터 디즈니 팬입니다ㅋㅋㅋ", 
                "디즈니 의 새로운 해석도 놀랍고. 윌스미스도 신의한수!"]

for review in review_list1: # review가 review_list1을 반복
    print(review.replace("ㅋ","")) # "ㅋ"이 들어간 문자를 " " 공백으로 바꿔서 출력
for review in review_list2: # review가 review_list2를 반복
    print(review.replace("ㅋ","")) 
```

```python
def string_replace(list_test): # 함수 생성
    for review in list_test: # review가 list_test를 반복
        print(review.replace("ㅋ","")) 
```

```python
string_replace(review_list1) # 생성한 함수를 review_list1 리스트에 적용해 함수를 실행
```

```python
# 두 정수와 연산기호를 입력하여 입력한 연산자에 맞게 계산해주는 함수 생성
def num_cal(a,b,c):
    if b == '+':
        result = a+c
    elif b == '-':
        result = a-c
    elif b == '*':
        result = a*c
    elif b == '/':
        result = a/c
    elif b == '%':
        result = a%c
    elif b == '//':
        result = a//c
    return result    
```

### 15. 가변인자함수 ( *args )
- 함수의 매개변수 숫자를 모를 때 사용하는 방식
- 함수의 매개변수를 몇개를 입력하든 함수 내에서 튜플로 인식
- 일반 매개 변수와 가변 인자 함수를 같이 사용하는 것도 가능

```python
def add(*args):
    sum = 0 
    for num in args: #num이 args를 반복
        sum += num #sum에 num로 들어온 숫자들을 모두 누적해서 더함.
    return sum 
```

```python
add(1,2,3)
```

### 16. 키워드 함수 ( **kwargs )
- 일반매개변수 나 가변인자함수 같이 사용이 가능
- 함수 내에서 딕셔너리 형태로 인식
- 키워드 인자 + 가변인자함수 또는 키워드 인자 + 일반매개변수 사용시 입력 순서는 <br>무조건 키워드인자 함수는 뒤에 나와야 한다.

```python
def print_map(**kwargs):
    print(kwargs)
    for key, value in kwargs.items(): # key와 value가 kwargs의 요소값들을 반복
        print(key, "/", value)
```

```python
print_map(하나=1)
```
